# **Forming Anomaly Detection Models to capture the bankruptcy cases in the Dataset:-** 

## **Building Isolation Forest Model in UnSupervised Setting:-**




```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
from sklearn.metrics import roc_auc_score
```

```python
clf=IsolationForest(n_estimators=100, max_samples='auto', \
                        max_features=1.0, bootstrap=False, n_jobs=-1, random_state=0, verbose=0, contamination=0.033)
clf.fit(train_X)
```

```python
pred_train = clf.predict(train_X)
np.unique(pred_train, return_counts=True)
```

```python
target_names = ['bankrupt_company', 'normal_company']
print(classification_report(train_Y, pred_train, target_names=target_names))
print ("AUC: ", "{:.1%}".format(roc_auc_score(train_Y, pred_train)))
cm = confusion_matrix(train_Y, pred_train )
print(cm)
```

```python
pred_test = clf.predict(test_X)
np.unique(pred_test, return_counts=True)
```

```python
target_names2 = ['bankrupt_company', 'normal_company']
print(classification_report(test_Y, pred_test, target_names=target_names2))
print ("AUC: ", "{:.1%}".format(roc_auc_score(test_Y, pred_test)))
cm2 = confusion_matrix(test_Y, pred_test)
print(cm2)
```

### In Isolation Forest, we are able to get close to 64% AUC score along with 33% recall on test set which is not bad considering the level of imbalance in the dataset. 

## **Building Local Outlier Factor Model in UnSupervised Setting:-**

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
clf2 = LocalOutlierFactor(n_neighbors=35, contamination=0.03, novelty=False)
clf2.fit(train_X)
```

```python
target_names = ['bankrupt_company', 'normal_company']
print(classification_report(train_Y, pred_train2, target_names=target_names))
print ("AUC: ", "{:.1%}".format(roc_auc_score(train_Y, pred_train2)))
cm_lof = confusion_matrix(train_Y, pred_train2 )
print(cm_lof)
```

```python
pred_test2 = clf.fit_predict(test_X)
np.unique(pred_test2, return_counts=True)
```

```python
target_names = ['bankrupt_company', 'normal_company']
print(classification_report(test_Y, pred_test2, target_names=target_names))
print ("AUC: ", "{:.1%}".format(roc_auc_score(test_Y, pred_test2)))
cm2 = confusion_matrix(test_Y, pred_test2)
print(cm2)
```

## **Different Techniques for Doing Feature Selection:-**

## 1. Using p_values derived from OLS

```python
#scaling the data
scaler=StandardScaler()
X_2=scaler.fit_transform(X)
```

```python
import statsmodels.api as sm
model = sm.OLS(Y, X_2)
results = model.fit()
print(results.summary())
```

```python
p_val_ols= pd.DataFrame(zip(X.columns,results.pvalues),columns=['column','p_values'])
p_val_ols.head(5)
```

### **Segregating Features based on 95% confidence interval:-**

```python
# now let us choose the features which provide us 95% confidence (p-value<0.05)
columns_based_on_p_values_ols=[x for x in p_val_ols[p_val_ols.p_values<=0.05]['column']]
len(columns_based_on_p_values_ols)
```

```python
columns_based_on_p_values_ols
```

### **We reduced our columns to 19 from 94 using p_vlaues of OLS model.**
